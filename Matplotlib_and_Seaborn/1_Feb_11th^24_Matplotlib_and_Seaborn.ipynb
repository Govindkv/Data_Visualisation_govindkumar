```python
plt.plot(x,np.sin(x-0),color="blue")
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.grid()
plt.title("This is sin plot before real dataset execution to undersatand basic syntax")
plt.axis("equal")
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color="blue")
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.grid()
plt.title("This is sin plot before real dataset execution to undersatand basic syntax")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.axis("equal")
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color="blue",label="Sinx")
plt.plot(x,np.cos(x-0),color="red",label="Cosx")
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.grid()
plt.title("This is sin plot before real dataset execution to undersatand basic syntax")
plt.xlabel("X axis")
plt.ylabel("Y axis")
#plt.axis("equal")
plt.legend()
plt.show()
```

```python
#Histogram
#For continuous data you will make Histogram or Distribution Plot 
#For descrete data you make Bar plot
data=np.random.randn(1000) #10 bins
plt.hist(data,bins=20,histtype="step", color="g")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6,2))
plt.scatter(x,y,c="r",alpha=.9,marker="X")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plot")
plt.show()
```

```python
#Pie Chart
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python

my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = my_explode, shadow =  True)
plt.title("Pie chart")
plt.show()
```
