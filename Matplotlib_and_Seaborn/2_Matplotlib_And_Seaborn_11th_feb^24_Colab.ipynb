```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

#pie chart

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = my_explode, shadow =  True)
plt.title("Pie chart")
plt.show()
```

```python
# 3 d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c = "#FF5733", alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
